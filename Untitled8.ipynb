```python
import numpy as np
import matplotlib.pyplot as plt
def normal_density(mean, variance, x): 
    #calculate the standard deviation from the variance
    std_dev = np.sqrt(variance)

    #calculate the exponent term in the normal distribution formula
    exponent = -((x - mean) ** 2) / (2 * variance)

#calculate the normal density using the formula 
    density = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(exponent)

    return density

# values for mean 
mean_values = [0, 2, -1]
variance_values = [1, 0.5, 2]

#values for x to plot the density function
x_values = np.linspace(-5, 5, 1000)

# plotting for each combination of mean and variance
plt.figure (figsize=(10, 6))

for mean, variance in zip(mean_values, variance_values):
    density_values = normal_density(mean, variance, x_values)
    label = f'mean={mean}, variance={variance}' 
    plt.plot(x_values, density_values, label=label)


plt.title('Normal Density Function for Different µ and σ^2')
plt.xlabel('x')
plt.ylabel('fX(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scripy.integrate import quad

def normal_density(mean, variance, x):
    std_dev = np.sqrt(variance)
    exponent = -((x - mean) ** 2) / (2 * variance)
    density = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(exponent)
    return density

def integration(mean, variance, a, b):
    integrand = lambda x: normal_density(mean, variance, x)
    result, error = quad(integrand, a, b)
    return result

#given parameters for male heights
mean_height = 171 # µ in cm
variance_height = 7.1

#height range for probability calculation
lower_limit = 162
upper_limit = 190


probability = integration(mean_height, variance_height, lower_limit, upper_limit)

print(f"The probability that a randomly chosen person has a height between {lower_limit}cm and {upper_limit}cm is approximately {probability:.4f}")
```
